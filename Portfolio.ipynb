```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('languages.csv')
df.m = pd.to_datetime(df.m)
df.rename(columns={'m': 'Date', 'TagName': 'Tag', 'Unnamed: 2': 'Number of Posts'}, inplace=True)
```

```python
tags_to_remove = ['assembly', 'delphi', 'perl', 'php']
df = df[~df.Tag.isin(tags_to_remove)]
```

```python
reshaped_df = df.pivot(index='Date', columns='Tag', values='Number of Posts')
reshaped_df.fillna(0, inplace=True)

plt.figure(figsize=(16,10))
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('Date', fontsize = 14)
plt.ylabel('Number of Posts', fontsize = 14)
plt.ylim(0, 35000)

for column in reshaped_df.columns:
  plt.plot(reshaped_df.index, reshaped_df[column], linewidth=3,
           label=reshaped_df[column].name)
plt.legend(fontsize=16)
```

```python
df_data = pd.read_csv('nobel_prize_data.csv')
```

```python
top10birth_countries = df_data.birth_country.value_counts().head(10)
top10birth_countries
```

```python
h_bar = px.bar(x=top10birth_countries.index,
               y=top10birth_countries.values,
               orientation='h',
               color=top20_countries.prize,
               color_continuous_scale='Viridis',
               title='Top 20 Countries by Number of Prizes')

h_bar.update_layout(xaxis_title='Number of Prizes',
                    yaxis_title='Country',
                    coloraxis_showscale=False)
h_bar.show()
```
